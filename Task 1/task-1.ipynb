### **Gerder distribution by this ratio**

##### **It is estimated to be about 1.10 or 1.06 or within a narrow range from 1.03 to 1.06 males per female. The sex ratio for the entire world population is approximately 101 males to 100 females.**

```python
# Calculate male and female populations based on the global sex ratio
df['Male Population'] = df['2023 population'] * (101 / (101 + 100))
df['Female Population'] = df['2023 population'] * (100 / (101 + 100))

# Summing male and female populations for the bar plot
gender_counts = {
    'Male Population': df['Male Population'].sum(),
    'Female Population': df['Female Population'].sum()
}
```

```python

# Plotting the gender distribution
plt.figure(figsize=(4, 5))
sns.barplot(x=list(gender_counts.keys()), y=list(gender_counts.values()), palette=['blue', 'pink'])
plt.title('Gender Distribution in Global Population')
plt.xlabel('Gender')
plt.ylabel('Population')
plt.show()
```

#### **AGE Distribution**

```python
np.random.seed(42)
n_samples = 1000
ages = np.random.normal(35, 15, n_samples)
ages = np.clip(ages, 0, 90).astype(int)  # Clip ages between 0 and 90

# Create a DataFrame for age data
age_df = pd.DataFrame({'Age': ages})
```

```python
plt.figure(figsize=(5, 5))
sns.histplot(age_df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution in the Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
